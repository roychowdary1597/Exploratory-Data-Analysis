```python
Total_Price = []

# Loop through each car
for car in soup.find_all("div", class_="list-item"):
    # Look for price tag under any of the 3 known classes
    price_tag = (
        car.find("div", class_="price_buy current") or
        car.find("div", class_="price_buy upcoming") or
        car.find("div", class_="price_buy archive")
    )

    if price_tag:
        price = price_tag.get_text(strip=True)
    else:
        price = "Not available"

    Total_Price.append(price)

# Output the results
print("Total prices found:", len(Total_Price))
print("Prices:", Total_Price)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://ev-database.org/uk/#group=vehicle-group&p=0-50"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Store seating capacities
seating_list = []

# Loop through each car listing
for car in soup.find_all("div", class_="list-item"):
    # Find the span with the tooltip for "Number of seats"
    seat_tag = car.find("span", attrs={"data-tooltip": "Number of seats"})
    if seat_tag:
        seat_number = seat_tag.find("span")
        if seat_number:
            seating_list.append(seat_number.text.strip())
        else:
            seating_list.append("Not available")
    else:
        seating_list.append("Not available")

# Print results
print("Total cars:", len(seating_list))
print("Seating capacities:", seating_list)
```
